Handling Null Values

```python
credits_df.isnull().sum()
```

```python
credits_df['checking_balance'].replace('unknown',np.nan,inplace=True)
credits_df['checking_balance'].replace('< 0 DM',0,inplace=True)
credits_df['checking_balance'].replace('1 - 200 DM',199/2,inplace=True)
credits_df['checking_balance'].replace('> 200 DM',200,inplace=True)
```

```python
credits_df['savings_balance'].value_counts()
```

```python
credits_df['savings_balance'].replace('unknown',np.nan,inplace=True)
credits_df['savings_balance'].replace('< 100 DM',100,inplace=True)
credits_df['savings_balance'].replace('100 - 500 DM',200,inplace=True)
credits_df['savings_balance'].replace('500 - 1000 DM',250,inplace=True)
credits_df['savings_balance'].replace('> 1000 DM',1000,inplace=True)
```

```python
credits_df['employment_duration'].value_counts()
```

```python
credits_df['employment_duration'].replace('1 - 4 years',4,inplace=True)
credits_df['employment_duration'].replace('> 7 years',8,inplace=True)
credits_df['employment_duration'].replace('4 - 7 years',7,inplace=True)
credits_df['employment_duration'].replace('< 1 year',1,inplace=True)
credits_df['employment_duration'].replace('unemployed',0,inplace=True)
```

```python
credits_df.head()
```

```python
mode1=credits_df['checking_balance'].mode().iloc[0]
mode2=credits_df['savings_balance'].mode().iloc[0]
credits_df['checking_balance'].fillna(mode1,inplace=True)
credits_df['savings_balance'].fillna(mode2,inplace=True)
```

```python
credits_df['zscore1']=(credits_df['amount']-credits_df['amount'].mean())/credits_df['amount'].std()
credits_df.drop(credits_df[np.abs(credits_df['zscore1'])>3].index, inplace = True)
credits_df['zscore2']=(credits_df['age']-credits_df['age'].mean())/credits_df['age'].std()
credits_df.drop(credits_df[np.abs(credits_df['zscore2'])>3].index, inplace = True)
```

```python
credits_df.drop(['zscore1','zscore2'],inplace=True,axis=1)
```

```python
x=credits_df.drop(['default'],axis=1)
y=credits_df[['default']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

LOGISTIC REGRESSION

```python
logistic_regression=LogisticRegression()
logistic_regression.fit(x_train,y_train)
print(logistic_regression.score(x_train,y_train),logistic_regression.score(x_test,y_test))
```

```python
predictions=logistic_regression.predict(x_test)
print(metrics.classification_report(y_test,predictions))
```

KNN

```python
knn=KNeighborsClassifier(n_neighbors=80,p=2) #p=1, manhattan p=2, euclidian
knn.fit(x_train,y_train)
print(knn.score(x_train,y_train),knn.score(x_test,y_test))
```

```python
predictions=knn.predict(x_test)
print(metrics.classification_report(y_test,predictions))
```

SVC

```python
clf=SVC(kernel='linear',C=2,gamma='auto')
clf.fit(x_train,y_train)
print(clf.score(x_train,y_train),clf.score(x_test,y_test))
```

```python
predictions=clf.predict(x_test)
print(metrics.classification_report(y_test,predictions))
```

ADABOOSTING

```python
adb=AdaBoostClassifier()
adb.fit(x_train,y_train)
print(adb.score(x_train,y_train),adb.score(x_test,y_test))
```

```python
predictions=adb.predict(x_test)
print(metrics.classification_report(y_test,predictions))
```

GRADIENTBOOSTING

```python
gbc=GradientBoostingClassifier(n_estimators=30)
gbc.fit(x_train,y_train)
print(gbc.score(x_train,y_train),gbc.score(x_test,y_test))
```

```python
predictions=gbc.predict(x_test)
print(metrics.classification_report(y_test,predictions))
```

DESICIONTREE

```python
dtm=DecisionTreeClassifier(max_depth=0.5,criterion='entropy')
dtm.fit(x_train,y_train)
print(dtm.score(x_train,y_train),dtm.score(x_test,y_test))
```

```python
predictions=dtm.predict(x_test)
print(metrics.classification_report(y_test,predictions))
```

RANDOM FOREST

```python
rfc=RandomForestClassifier(n_estimators=40,max_features=9)
rfc.fit(x_train,y_train)
print(rfc.score(x_train,y_train),rfc.score(x_test,y_test))
```

```python
predictions=rfc.predict(x_test)
print(metrics.classification_report(y_test,predictions))
```

BAGGING CLASSIFIER

```python
bgc1=BaggingClassifier(n_estimators=12,base_estimator=dtm)
bgc1.fit(x_train,y_train)
print(bgc1.score(x_train,y_train),bgc1.score(x_test,y_test))
```

```python
predictions=bgc1.predict(x_test)
print(metrics.classification_report(y_test,predictions))
```

Gradient boosting is far better when compared to other models since the train score and test score are less deviating.
